```python
# リスト5.9.1：ファイルの出力
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
dat = [1, 3, 2]
ax.plot(dat)
fname = 'save_fig.svg'
plt.savefig(fname)
plt.show()
```

```python
# リスト5.9.2：出力時にフィギュアオブジェクトの書式を設定
fig = plt.figure()
ax = fig.add_subplot(111)
dat = [1, 3, 2]
ax.plot(dat)
fname = 'mpl_save_fig_prop_setting.svg'
plt.savefig(fname, dpi=300, facecolor='grey',
            edgecolor='k', linewidth=10,
            frameon=True, tihgt_layout=False)
plt.show()
```
